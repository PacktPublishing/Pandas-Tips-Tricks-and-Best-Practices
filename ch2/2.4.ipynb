retrieval using query

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('factbook.db')
query = "SELECT name, area, population FROM facts"
df = pd.read_sql(query, conn)
conn.close()
df
```

retrieval using SQLAlchemy 

```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine('sqlite:///factbook.db') 
query = "SELECT name, area, population FROM facts"
df = pd.read_sql(query, engine)

df
```

read entire table using table name for factbook.db

```python
df = pd.read_sql_table('facts', 'sqlite:///factbook.db')
df
```

read table equivalent

```python
conn = sqlite3.connect('factbook.db')
query = "SELECT * FROM facts"
df = pd.read_sql(query, conn)
conn.close()
df
```

chunksize demo

```python
dfl = []
for dfc in pd.read_sql('Observation', 'sqlite:///diamonds.db', chunksize=10_000):
    # process dfc here
    dfl.append(dfc)
    
df = pd.concat(dfl)
```
