read direct from JSON output of an API

```python
import pandas as pd
from urllib.parse import urlencode

ep = "https://aviationweather.gov/api/data/airport"
params = {'ids': 'KSFO,KLAX,KOAK,KSJO,KAPC,KCCR,KSQL', 'format': 'json'}
url = ep + "?" + urlencode(params)
pd.read_json(url)
```

Often, JSON output needs to be parsed first

```python
import pandas as pd
from urllib.parse import urlencode

url = "https://openlibrary.org/search.json"
params = {'author': 'Hemingway', 'sort': 'new'}
url = url + "?" + urlencode(params)
pd.read_json(url)
```

store the JSON response in a file (ol.json) so we can examine it

```python
import pandas as pd
import requests
import json

url = "https://openlibrary.org/search.json"
params = {'author': 'Hemingway', 'sort': 'new'}

js = requests.get(url, params=params).json()
with open('ol.json', 'w') as f:
    json.dump(js, f)
```

Using requests to retrieve a JSON response

```python
import pandas as pd
import requests

url = "https://openlibrary.org/search.json"
params = {'author': 'Hemingway', 'sort': 'new'}

js = requests.get(url, params=params).json()
pd.DataFrame(js['docs'])
```
