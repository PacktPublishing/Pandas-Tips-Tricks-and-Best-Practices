```python
import numpy as np
import pandas as pd

df = pd.read_csv('Airbnb_Open_Data_ch5.csv')
df.info()
```

indexing operator selection

```python
df[['name', 'neighborhood', 'price']].head()
```

loc attribute selection

```python
df.loc[:,['name', 'neighborhood', 'price']].head()
```

loc attribute slice range - note inclusive

```python
df.loc[:,'name':'neighborhood'].head()
```

iloc attribute selection

```python
df.iloc[:, [1, 6, 13]].head()
```

iloc attribute can accept a slice

```python
df.iloc[:, 1:7].head()
```

filter method

```python
df.filter(['name', 'neighborhood', 'price']).head()
```

partial column label matching

```python
df.filter(like='host').head()
```

```python
df.filter(regex='^host').head()
```

Generate a DataFrame with 100 columns

```python
rows = 10
cols = 100

columns = [f'col{i}' for i in range(1, cols + 1)]
data = np.arange(1, (rows * cols) + 1).reshape(rows, cols)
df = pd.DataFrame(data, columns=columns)
df
```

Filter only columns that match a given regular expression pattern (col70 to col88, even only)

```python
df.filter(regex='col[78][0,2,4,6,8]').head()
```
