```python
import pandas as pd

df = pd.read_csv('Airbnb_Open_Data_ch5.csv')
```

Get rows from a slice index 

```python
df.loc[10:12, ['id', 'host_id', 'name']]
```

Get rows from a slice position - stop value is exclusive with iloc

```python
df.iloc[10:12, [0, 2, 1]]
```

The following includes the stop value from the list of positions

```python
df.iloc[[10, 11, 12], [0, 2, 1]]
```

Custom indexes

Set host_id as the index

```python
print(df.host_id.is_unique)
df = df.set_index('host_id')
df = df.sort_index()
df.index.is_monotonic_increasing
```

Print duplicate indexes

```python
df = df.drop_duplicates()
dupe_indexes = df[df.index.duplicated()].index
df.loc[dupe_indexes]
```

update the index and verify its uniqueness

```python
df = df.reset_index()
new_id = df.host_id.max() + 1
idx_change = df[df.host_id == 29531702698].index[0]
df.loc[idx_change, 'host_id'] = new_id
df.index.is_unique
```

Hierarchical Index

Set to a multiindex of neighborhood_group and neighbor

```python
df = df.set_index(['neighborhood_group', 'neighborhood'])
```

Selecting from this DataFrame using loc (specify values for each index) - results in a warning

```python
df.loc["Brooklyn", "Cobble Hill"]
```

Get rid of the warning by sorting the index. Null values must be sorted first, otherwise the warning will persist.

```python
df = df.sort_index(na_position='first')
```
